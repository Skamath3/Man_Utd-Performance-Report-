```python
#Testing Method

#function to stabilize data
def get_stationarity(timeseries):

#Rolling statistics
  rolling_mean = timeseries.rolling(window=8).mean()
  rolling_std = timeseries.rolling(window=8).std()

#Dickey-Fuller test:
  result = adfuller(timeseries['Close'])
  print('ADF Statistic: {}'.format(result[0]))
  print('p value: {}'.format(result[1]))
  print('Critical Values:')
  for key,value in result[4].items():
    print('\t{}: {}'.format(key,value))
```

```python
#Results of Testing Method
Stocks_log_shift = Stocks_log - Stocks_log.shift() 
Stocks_log_shift.dropna(inplace=True)
get_stationarity(Stocks_log_shift)
```

```python
Stocks_log.isnull().values.any() #FindingNullValues
```

```python
decomposition = seasonal_decompose(Stocks_log, freq= 52)
model = ARIMA(Stocks_log, order=(4,1,2))
results = model.fit(disp=0)
results
```

```python
#Series for Prediction

predictions_ARIMA_diff = pd.Series(results.fittedvalues, copy = True)
predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()
```

```python
predictions_ARIMA_log = pd.Series(Stocks_log['Close'].iloc[0], index=Stocks_log.index)
predictions_ARIMA_log = predictions_ARIMA_log.add(predictions_ARIMA_diff_cumsum,fill_value=0)
predictions_ARIMA = np.exp(predictions_ARIMA_log)
```
